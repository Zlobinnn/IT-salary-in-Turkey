# Основная информация о датасете
Набор данных имеет следующие атрибуты:
- Город трудоустройства (Şehir)
- Уровень опыта (Seviye)
- Позиция (Pozisyon)
- Средняя зарплата (Maaş_Net_TL)

В каждом столбце 1.000 значений

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dataset.csv")
df.head()
```

```python
print("Размер:", df.shape)
df.info()
df.describe()
df.isna().sum()
```

# Средняя зарплата по городам

Исходя из графика видно, что разница в зарплатах между различными городами незначительна

```python
city_salary = df.groupby("Şehir")["Maaş_Net_TL"].mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(
    x=city_salary.values,
    y=city_salary.index,
    hue=city_salary.index,
    palette="viridis",
    legend=False
)
plt.title("Средняя зарплата по городам")
plt.xlabel("Средняя зарплата (Турецкая Лира)")
plt.ylabel("Город")
plt.show()
```

# Зарплата по уровням опыта

Из графиков видно, что средняя зарплата Middle-разработчика на 56% выше, чем у Junior, а у Senior на 52% выше, чем у Middle

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x="Seviye",
    y="Maaş_Net_TL",
    hue="Seviye",
    data=df,
    palette="mako",
    legend=False
)
plt.title("Зарплата по уровням опыта")
plt.xlabel("Уровень")
plt.ylabel("Зарплата (Турецкая Лира)")
plt.show()

salary_by_level = df.groupby("Seviye")["Maaş_Net_TL"].mean().sort_values()
display(salary_by_level)

percent_diff = salary_by_level.pct_change() * 100

print("\nПроцентное различие между уровнями:")
for i in range(1, len(salary_by_level)):
    prev_level = salary_by_level.index[i-1]
    curr_level = salary_by_level.index[i]
    diff = percent_diff.iloc[i]
    print(f"От {prev_level} к {curr_level}: {diff:.2f}%")
```

# Средняя зарплата по позициям

Исходя из графика видно, что **лидером** в заработной плате среди других позиций является **Data Engineer со значением 106 тыс. Турецких Лир**.
Остальные 4 наиболее высокооплачиваемых позиции:
- Data Science: 90,417 TL
- Machine Learning: 88,915 TL
- Embedded Software Developer: 87,000 TL
- Cybersecurity: 84,714 TL

```python
position_salary = df.groupby("Pozisyon")["Maaş_Net_TL"].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(
    x=position_salary.values,
    y=position_salary.index,
    hue=position_salary.index,
    palette="cubehelix",
    legend=False
)
plt.title("Средняя зарплата по позициям")
plt.xlabel("Средняя зарплата (Турецкая Лира)")
plt.ylabel("Позиция")
plt.show()

for i, (pos, sal) in enumerate(position_salary.head(5).items(), start=1):
    print(f"{i}. {pos}: {sal:,.0f} TL")
```

# Зарплата по уровням и городам

Исходя из графиков видно, что для Junior и Middle-разработчиков зарплата не сильно зависит от города.
Однако, для Senior разница наглядна:
Лидирует Анкара, затем Стамбул и Измир и с большим отрывом все остальные города

```python
plt.figure(figsize=(12,6))
sns.barplot(
    x="Seviye",
    y="Maaş_Net_TL",
    hue="Şehir",
    data=df,
    errorbar=None
)
plt.title("Зарплата по уровням и городам")
plt.ylabel("Зарплата (Турецкая Лира)")
plt.xlabel("Уровень опыта")
plt.legend(title="Город", bbox_to_anchor=(1.05, 1))
plt.show()
```
